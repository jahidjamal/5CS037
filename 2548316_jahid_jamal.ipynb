HDI Category

```python
def hdi_category(x):
    if x < 0.55:
        return "Low"
    elif x < 0.70:
        return "Medium"
    elif x < 0.80:
        return "High"
    else:
        return "Very High"

hdi_2022_df["HDI Category"] = hdi_2022_df["hdi"].apply(hdi_category)
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
```

PROBLEM 2 – SOUTH ASIA

```python
south = ["Afghanistan","Bangladesh","Bhutan","India","Maldives","Nepal","Pakistan","Sri Lanka"]
sa = df[df["country"].isin(south)].copy()
sa.to_csv("HDI_SouthAsia.csv", index=False)
```

```python
# Create Composite Score properly
sa["Composite"] = 0.3 * sa["life_expectancy"] + 0.3 * sa["gross_inc_percap"]
```

```python
# Get 2022 data for South Asia
sa_2022 = sa[sa["year"] == 2022]

# Compute average Composite Score by country
composite_avg = sa_2022.groupby("country")["Composite"].mean().sort_values(ascending=False)

# Select top 5
top5 = composite_avg.head(5)

# Plot
plt.figure(figsize=(8,5))
top5.plot(kind="bar")
plt.title("Top 5 South Asian Countries by Composite Development Score")
plt.xlabel("Country")
plt.ylabel("Composite Score")
plt.show()
```

```python
# Filter South Asia for 2022
sa_2022 = sa[sa["year"] == 2022]

# Rank by HDI
hdi_rank = sa_2022.groupby("country")["hdi"].mean().rank(ascending=False)

# Rank by Composite Score
composite_rank = sa_2022.groupby("country")["Composite"].mean().rank(ascending=False)

# Build Table 3
table3 = pd.DataFrame({
    "country": hdi_rank.index,
    "HDI_Rank": hdi_rank.values,
    "Composite_Rank": composite_rank.values
}).sort_values("HDI_Rank")

table3
```

```python
sa_2022 = sa[sa["year"]==2022].copy()

Q1 = sa_2022["hdi"].quantile(0.25)
Q3 = sa_2022["hdi"].quantile(0.75)
IQR = Q3 - Q1

sa_2022["Outlier"] = (sa_2022["hdi"] < Q1 - 1.5*IQR) | (sa_2022["hdi"] > Q3 + 1.5*IQR)

table4 = sa_2022[["country","hdi","gross_inc_percap","Outlier"]]
table4
```

```python
def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]

sa_outliers = detect_outliers(p1b_sa, "hdi")
me_outliers = detect_outliers(p1b_me, "hdi")

sa_outliers[["country","year","hdi"]]
me_outliers[["country","year","hdi"]]
```

```python
p1b_sa["Outlier"] = p1b_sa.index.isin(sa_outliers.index)
sns.scatterplot(data=p1b_sa, x="gross_inc_percap", y="hdi", hue="Outlier")
plt.title("South Asia Outliers")
plt.show()

p1b_me["Outlier"] = p1b_me.index.isin(me_outliers.index)
sns.scatterplot(data=p1b_me, x="gross_inc_percap", y="hdi", hue="Outlier")
plt.title("Middle East Outliers")
plt.show()
```
